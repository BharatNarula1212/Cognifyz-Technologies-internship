#### IMPORT ALL THE REQUIRED LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
import warnings
warnings.filterwarnings("ignore")
```

#### LOAD DATASET

```python
df=pd.read_csv("Dataset  (1).csv")
```

```python
df.head()
```

#### Remove rows with missing values in the 'Cuisines' column

```python
df=df.dropna(subset=['Cuisines'])
```

#### Identify the most common combinations ofcuisines in the dataset.

```python
cuisine_combinations=df['Cuisines'].str.split(', ').apply(sorted).apply(', '.join)
most_common_cuisine_combinations=cuisine_combinations.value_counts().head(5)
print("Most Common Cuisine Combinations:")
print(most_common_cuisine_combinations)
```

#### Determine if certain cuisine combinationstend to have higher ratings.

```python
average_ratings_by_combination=df.groupby(cuisine_combinations)['Aggregate rating'].mean()
print("\nAverage Ratings by cuisine combination:")
print(average_ratings_by_combination)

print('-'*100)

overall_average_rating=df['Aggregate rating'].mean()
print("\nOverall Average Rating:",overall_average_rating)

print('-'*100)

for combinations,avg_rating in average_ratings_by_combination.items():
    print(f"Cuisine Combiantions:{combinations},Average Rating: {avg_rating:.2f},Difference from Overall Average:{avg_rating-overall_average_rating:.2f}")
```

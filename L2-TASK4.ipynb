#### IMPORT ALL THE REQUIRED LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
import warnings
warnings.filterwarnings("ignore")
```

#### LOAD DATASET

```python
df=pd.read_csv("Dataset  (1).csv")
```

```python
df.head()
```

#### Identify if there are any restaurant chains present in the dataset.

```python
# Identify restaurant chains based on the 'Restaurant Name'
restaurant_chains = df['Restaurant Name'].value_counts()

# Filter out individual restaurants (considering chains with multiple locations)
chains = restaurant_chains[restaurant_chains > 1].index

# Create a DataFrame for restaurant chains
chain_df = df[df['Restaurant Name'].isin(chains)].copy()
chain_df.head()
```

#### Analyze the ratings and popularity of different restaurant chains.

```python
chain_analysis=chain_df.groupby('Restaurant Name').agg({
    'Aggregate rating':'mean',
    'Votes':'sum',
    'Average Cost for two':'mean'
}).sort_values(by='Votes',ascending=False)

# Display the analysis
print("Restaurant Chains Analysis:")
print(chain_analysis)
```
